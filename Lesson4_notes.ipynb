## 练习

编写函数计算n的阶乘

然后调用脚本根据命令行的输入计算阶乘

```python
import sys
x = sys.argv[1]  # sys.argv得到的是字符串列表
y = 1

if x % 1 != 0:
    print("我想你是输错了")
else:
    for i in range(1,x+1):  # 注意，阶乘必须从1开始！
        y = y * i

print('{}的阶乘为{}'.format(x,y))
```

## 练习：会合时间

已知大约780天发生一次火星冲日。2014年4月19日发生了火星冲日，求下一次火星冲日是什么时候？

```python
import datetime
from datetime import datetime as dtdt
d = dtdt.strptime('2014/4/9','%Y/%m/%d') + datetime.timedelta(780) # path 
# 注意区分strftime从时间到字符串 和strptime从字符串到时间
print("下一次火星冲日的时间为：" + str(d))
```

## 练习：闰月设置

Ty回归年=365.242190天，Tm朔望月=29.530588天。计算1000年内，误差最小的闰月周期

```python
import numpy as np

year = 1000
Ty = 365.242190
Tm = 29.530588

best_error = float('inf')
best_N = 0
best_M = 0

for N in range(1, year + 1):
    # 理论上最合适的闰月数
    M = round(N * (Ty / Tm - 12))  # 找到年已确定下的应当插入的月份，并且注意是四舍五入的
    
    # 实际误差
    error = abs(N * Ty - (12 * N + M) * Tm)
    
    if error < best_error:
        best_error = error
        best_N = N
        best_M = M

print("最佳闰月周期：")
print(f"N = {best_N} 年")
print(f"M = {best_M} 个闰月")
print(f"误差 = {best_error:.6f} 天")
```

## 作业：批量创建文件

在工作目录下创建文件夹1Kfiles
在其中生成1000个文件，文件名格式为
test001.dat,test002.dat,...,test999.dat
文件内容有两行：
第一行为文件生成的时间和日期
第二行为文件序号的平方根，要保留20位有效数字

```python
# 导入模块
import datetime as dt
from datetime import datetime as dtdt
from decimal import Decimal as D
from decimal import getcontext
# 设置精度
getcontext().prec=20
```

```python
# 文件根目录
file_dir = "D:\\my study\\Python Study\\Lesson\\1Kfiles\\"
# 创建文件并写入
for i in range(1,1001):
    filename = f"{i:03d}.dat"  # 规范文件名格式   # i表示要规范的变量，03d表示宽度为3，不足的在左边补零
    f = open(file_dir + filename,"w",encoding="utf-8")
    f.write(dtdt.strftime(dtdt.now(),'%Y/%m/%d %H:%M:%S.%f') + "\n")
    f.write(str(D(i).sqrt()) + "\n")
    f.close()
print("批量创建文件结束")
```
